## Parte 2: Aplicaciones en Python

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot #import matplotlib.pyplot as plt

img = Image.open("tarea2.color.jpg")
row, col =  img.size
data = np.zeros((row*col, 5))
pixels = img.load()

for i in range(1,row):
    for j in range(1,col):
        r,g,b =  pixels[i,j]
        data[i*col + j,:] = r,g,b,i,j
data
```

```python
img
```

```python
img = img.convert("LA")
```

### SVD

```python
A = np.matrix(data)
#A
```

```python
u, s, v = np.linalg.svd(A, full_matrices=1)
```

```python
print(u)
print("")
print(s)
print("")
print(v)
```

```python
matplotlib.pyplot.figure(figsize=(9,6))
matplotlib.pyplot.imshow(img, cmap='gray')
for k in range(3):
    B = np.matrix(u[:, :i]) * np.diag(s[:i]) * np.matrix(v[:i, :])
    matplotlib.pyplot.imshow(B, cmap='gray')
    matplotlib.pyplot.title("k = %s" % k)
    matplotlib.pyplot.show()
```
