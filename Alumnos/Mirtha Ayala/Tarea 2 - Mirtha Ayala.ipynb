```python
#1.	¿Por qué una matriz equivale a una transformación lineal entre espacios vectoriales? 
Función lineal es toda aplicación cuyo dominio y codominio sean espacios vectoriales que cumpla la siguiente definición:
Sean V y W espacios vectoriales sobre el mismo cuerpo K. Una transformación T de V en  W es una transformación lineal 
si para todo par de vectores u, v en V y para todo escalar k en K, se satisface que:
1.	Suma de vectores: T(u+v)=T(u)+T(v)
2.	Producto por escalar: T(ku)=kT(u)
Funciones de espacios vectoriales
```

```python
#2. ¿Cuál es el efecto de transformación lineal de una matriz diagonal y el de una matriz ortogonal?
Matriz Diagonal
Una matriz es diagonal y cuadrada cuando todas las entradas fuera de la diagonal principal son cero.
La regla de correspondencia está determinada por la diagonal, además las operaciones de suma y producto de matrices son
especialmente sencillas. 
Por ejemplo diag(a1,...,an) para una matriz diagonal que tiene las entradas a1,...,an en la diagonal principal, 
empezando en la esquina superior izquierda. Entonces, para la suma se tiene:
diag(a1,...,an) + diag(b1,...,bn) = diag(a1+b1,...,an+bn)
y para el producto de matrices,
diag(a1,...,an) • diag(b1,...,bn) = diag(a1b1,...,anbn).
La matriz diagonal diag(a1,...,an) es invertible si y sólo si las entradas a1,...,an son todas distintas de 0. 
En este caso, se tiene:
diag(a1,...,an)-1 = diag(a1-1,...,an-1).
Matriz Ortogonal
Una matriz es ortogonal (cuadrada y real) cuando se cumple lo siguiente: ATA=In
Es decir la inversa coincide con la transpuesta. Dadas las propiedades anteriores que caracterizan a dicha matriz, 
si queremos verlo como transformación lineal, estaríamos hablando de la matriz que multiplica al vector dado y da como 
resultado la identidad.
```

```python
#6. ¿Cómo interpretas la descomposición en valores singulares como una composición de tres tipos de transformaciones 
#lineales simples?
Geométricamente, la SVD significa que esferas en el dominio son transformadas en elipsoides en el codominio. 
Esencialmente se trata de reducir el rango de una matriz de r a k, donde k < r.
Toda transformación lineal es una rotación, redimensión de ejes canónicos y otra rotación. Por lo que la primera 
matriz se encarga de hacer una rotación, la segunda matriz (diagonal) contiene a los eigenvalores, por lo que se 
encarga de hacer una redimensión de los ejes canónicos, y la tercera matriz se encarda de dar una última rotación
```

```python
#7. ¿Qué relación hay entre la descomposición en valores singulares y la diagonalización?
Cualquier función lineal la puedes descomponer como función de tres matrices: dos funciones invertibles, en los 
extremos, y una diagonal (en medio) cuyos valores en la diagonal corresponden a los eigenvalores.
```

```python
#8. ¿Cómo se usa la descomposición en valores singulares para dar una aproximación de rango menor a una matriz?
Al hacer la descomposición, se tiene la matriz diagonal con los eigenvectores ordenados en columnas de mayor a menor 
importancia, por lo que basta tomar un número menor de esas columnas y renglones de las otras matrices para tener una 
aproximación a la matriz original
```

```python
#Recibir una imagen blanco y negro y aproximarla con la descomposición SVD
%matplotlib inline
import matplotlib.pyplot as plt

from PIL import Image
import numpy as np

img = Image.open("C:/Users/Public/Pictures/Sample Pictures/Koala.jpg")
imggray = img.convert('LA')
#img.show()

#plt.figure(figsize=(10, 8))
#plt.imshow(img)

imgmat = np.array(list(imggray.getdata(band=0)), float)
imgmat.shape = (imggray.size[1], imggray.size[0])
imgmat = np.matrix(imgmat)
#plt.figure(figsize=(9,6))
#plt.imshow(imgmat, cmap='gray');

U, sigma, V = np.linalg.svd(imgmat)

reconstimg = np.matrix(U[:, :1]) * np.diag(sigma[:1]) * np.matrix(V[:1, :])
plt.imshow(reconstimg, cmap='gray');


#Even with just the second and third vectors, the shape of the arch begins to appear.

for i in range(2, 4):
    reconstimg = np.matrix(U[:, :i]) * np.diag(sigma[:i]) * np.matrix(V[:i, :])
    plt.imshow(reconstimg, cmap='gray')
    title = "n = %s" % i
    plt.title(title)
    plt.show()

#The loop below shows the reconstructed image using the first  nn  vectors of the singular value decomposition 
#( nn  is shown in the title of the plot). The first 50 vectors produce an image very close the original image, 
#while taking up only  50∗3900+50+50∗26003900∗2600≈3.2%50∗3900+50+50∗26003900∗2600≈3.2% as much space as the 
#original data.

for i in range(120, 130, 5):
    reconstimg = np.matrix(U[:, :i]) * np.diag(sigma[:i]) * np.matrix(V[:i, :])
    plt.imshow(reconstimg, cmap='gray')
    title = "n = %s" % i
    plt.title(title)
    plt.show()
```
