## **1.- ¿Por qué una matriz equivale a una transformación lineal entre espacios vectoriales?**

### *R= Cuando una matriz $A$ multiplica a un vector $v$, lo transforma en otro vector dado $Av$. Ampliando dicha idea inicial se puede hablar de un conjunto de vectores v que pertenecen al espacio $V$ que al ser multiplicados, cada uno por $A$, son transformados. Existen ocasiones en que contamos con el producto de las transformaciones lineales de los espacios vectoriales, en este caso $V$, pero desconocemos la matriz A. En ese sentido, podemos asignar una matriz a cada transformación. Para vectores columna la entrada v pertenece a $V=R^n$ y la salida $T(v)$ pertenece a $Z=R^m$.   En ese sentido, La matriz $A$ para la transformación $A$, será de $m x n$. La selección de  bases en $V$ y $W$ determinarán la matriz $A$. Los vectores base estándar  de $R^n$ y $R^m$  son las columnas de la matriz identidad. Si escogemos dichos vectores tendremos una transformación de la forma $T(v)=Av$. Al mismo tiempo dichos espacios tienen otras bases posibles en la misma transformación representada por otras matrices.*                                  

### *Cuando V y W no son $R^n$ y $R^m$, siguen conteniendo bases. Cada elección determinada de base nos llevará a una matriz T. Cuando la base de entrada es diferente de la base de salida, la matriz para $T(v)=v$ no será la matriz identidad. En lugar de eso será la matriz que cambia las bases.*        

### *Cada v es una combinación única $c_1 v_1+c_2 v_2+⋯+c_n$ $v_n$ de la base de vectores de $v_i$. Si T es una transformación linear, $T(v)$ debe ser la misma combinación  $c_1 T(v_1)+c_2 T(v_2)+⋯+c_n T(v_n)$ de las salidas conocidas $T(v_i)$. *

### *Es decir una transformación representa el simil de una función. En ese sentido existe un regla de que asocia los valores de un conjunto dado (dominio para el caso de funciones) con otro conjunto (codominio para el caso de funciones también). Específicamente, de lo explicado anteriormente, existe una matriz que nos ayuda a relacionar ambos espacios vectoriales. Esa matriz, es la matriz A del planteamiento inicial en la respuesta.*

$$T:R^n→R^m$$
$$T(v ⃗ ):Av ⃗$$


## **2.- ¿Cuál es el efecto de transformación lineal de una matriz diagonal y el de una matriz ortogonal?**

### *R=De la misma manera que en la pregunta anterior, una transformación lineal es una “función” de matrices, donde a través de la multiplicación aplicada a un vector nos arroja como resultado un elemento en otro conjunto. Dicha multiplicación se denomina transformación lineal. En ese caso la matriz diagonal y la matriz ortogonal son matrices con propiedades particulares. En ese sentido, los resultados que arroja la multiplicación de la matriz A por el vector v son particulares. En ese sentido tenemos lo siguiente:*


### **Matriz Ortogonal**

### *Una matriz es ortogonal (cuadrada y real) cuando se cumple lo siguiente:*

$$A^{-1} = A^T$$
$$A^T A= I_n$$


### *Es decir la inversa coincide con la transpuesta. Dadas las propiedades anteriores que caracterizan a dicha matriz, si queremos verlo como transformación lineal, estaríamos hablando de la matriz que multiplica al vector dado y da como resultado la identidad. Adicionalmente, se puede agregar que dicha transformación lineal, según sea el caso, puede representar rotaciones, reflexiones o inversiones. *

### ** Matriz diagonal**

### *Una matriz es diagonal y cuadrada cuando todas las entradas fuera de la diagonal principal son cero. Cuando se multiplica esta matriz por cualquier matriz A del lado izquierdo los elementos de la i-esima fila de A se multiplican por los valores de la diagonal de la matriz diagonal. De la misma manera si se multiplica por el lado derecho los elementos j-esima fila de A se multiplican por los valores de la diagonal de la matriz diagonal.*

## ** 4.- ¿Qué es diagonalizar una matriz y que representan los eigenvectores?**

### *R= Es un proceso que parte de los eigenvectores y de los eigenvalores que se obtienen en la ecuación: $Ax= λx$. Supongamos que tenemos n eigenvectores linearmente independientes de A y que los acomodamos en las columnas de una matriz Z. Si multiplicamos $AZ$.*

### *Sería deseable separar las lambdas de las x del vector del resultado. En ese sentido, se hace el siguiente procedimiento:*

### *  Se puede observar que el resultado representa la matriz Z original y una matriz diagonal con los eigenvalores que se denominará $Λ$. Tenemos que $AZ=ZΛ$. Como se comentó desde un inicio los eigenvectores, que forman la matriz Z,  son linearmente independientes, lo que nos permite invertir nuestra matriz Z, quedando la diagonalización de la matriz de la siguiente manera: $Z^{-1} AZ=Λ$ o si se prefiere, $A=ZΛZ^{-1}$. Ampliando la ecuación anterior podemos encontrar lo siguiente( Es cierto que esta parte no tiene nada que ver con la respuesta, pero normalmente a partir de diagonalización se demuestran las potencias de una matriz):*

$$Ax= λx$$

$$A^2 x= λAx=λ^2 x$$

$$A^2=ZΛZ^{-1} ZΛZ^{-1}= ZΛ^2 Z^{-1}$$

$$A^n= ZΛ^n Z^{-1}$$

```python
import matplotlib.pyplot as plt
import numpy as np
import time

from PIL import Image
```

### * Para el procesamiento de la imagen se convertira el archivo a una matriz de numpy. Para esta operación en los comandos anteriores ya se había cargado el paquete numpy.*

```python
imggray = img.convert('LA')
plt.figure(figsize=(9, 6))
plt.imshow(imggray);
```

```python
imgmat = np.array(list(imggray.getdata(band=0)), float)
imgmat.shape = (imggray.size[1], imggray.size[0])
imgmat = np.matrix(imgmat)
plt.figure(figsize=(9,6))
plt.imshow(imgmat, cmap='gray');
```

```python
imgmat
```

### * En el comando imgmat podemos ver que la imagen ya se transformó en una matriz de numpy. Ahora para su procesamiento se aplicará el Descomposición en Valores Singulares de la siguiente manera:*

```python
U, sigma, V = np.linalg.svd(imgmat)
```

### * Ahora se puede procesar una aporximación a la imagen usando columnas de la matriz $U$ y los renglones de la matriz $V$. Entre más columnas y renglones agreguemos mayor será la nitidez de la imagen y mayor será su aproximación con la imagen original. Para esta parte del ejercicio iremos agregando columnas y renglones para ver el efecto de esta técnica.*

```python
reconstimg = np.matrix(U[:, :1]) * np.diag(sigma[:1]) * np.matrix(V[:1, :])
plt.imshow(reconstimg, cmap='gray');
```

```python
for i in range(2, 4):
    reconstimg = np.matrix(U[:, :i]) * np.diag(sigma[:i]) * np.matrix(V[:i, :])
    plt.imshow(reconstimg, cmap='gray')
    title = "n = %s" % i
    plt.title(title)
    plt.show()
```

### *Seguimos agregando...*

```python
for i in range(5, 100, 5):
    reconstimg = np.matrix(U[:, :i]) * np.diag(sigma[:i]) * np.matrix(V[:i, :])
    plt.imshow(reconstimg, cmap='gray')
    title = "n = %s" % i
    plt.title(title)
    plt.show()
```

## **¿Qué tiene que ver este proyecto con la compresión de imágenes?**

### * R= Como se ha observado a lo largo de este respuesta, la técnica de SVD nos permite comprimir imagenes mediante su traducción a una matriz, en este caso de grises. Entre más renglones y columnas agreguemos a nuestra imagen mayor será la matriz, pero también aumentará la calidad de la imagen. En ese sentido, se puede "comprimir" una imagen traduciéndola a una matriz que dependiendo de la necesidad de nitidez puedes agregar o quitar columnas para ajustar al proyecto en el que se esté usando. Como pudimos ver en el ejemplo, entre más renglones y columnas agregamos la matriz es más grande, pero también la calidad de la imagen.*
