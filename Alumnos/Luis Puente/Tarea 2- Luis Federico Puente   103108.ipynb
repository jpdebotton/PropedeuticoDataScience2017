### 1. ¿Por qué una matriz equivale a una trasformación linea entre espacios vectoriales?

Sea $A=\begin{bmatrix}a & b\\c & d\end{bmatrix}$

Y sea $T$ una transformación lineal de A tal que devuelve las columnas de la matriz 

$T(e_j)=Ae_j=A_j$


$T(e_1)=\begin{bmatrix}a & b\\c & d\end{bmatrix}\begin{bmatrix}1\\0\end{bmatrix}=\begin{bmatrix}a\\c\end{bmatrix}=A_1$

$T(e_2)=\begin{bmatrix}a & b\\c & d\end{bmatrix}\begin{bmatrix}0\\1\end{bmatrix}=\begin{bmatrix}b\\d\end{bmatrix}=A_2$

Entonces $A$ se puede escribir de la siguiente forma

$A=A(e_1+e_2)= Ae_1+Ae_2=A_1+A_2=T(e_1)+T(e_2)$

### 2. Cuál es el efecto de una transformación lineal de una matriz diagonal y el de una matriz ortogonal?

Sea $A=\begin{bmatrix}a & 0\\0 & d\end{bmatrix}$

Y sea $T:A\begin{bmatrix}x\\y\end{bmatrix}$ una transformación lineal de la matriz A 

Entones $\begin{bmatrix}a & 0\\0 & d\end{bmatrix}\begin{bmatrix}x\\y\end{bmatrix}=\begin{bmatrix}ax\\dy\end{bmatrix}$

Es decir, $T(x,y)=(ax,dy)$

Ahora, sea $A=\begin{bmatrix}a & b\\c & d\end{bmatrix}$ una matriz ortogonal

### 3. ¿Qué es la descomposición en valores singulares de una matriz?

La descomposición de valores singulares es una factorización de una matriz.
Formalmente:
$ A=U\Sigma\ V^T $

Con A una matriz de m*r, U matriz ortogonal de de m*m, 
V matriz ortogonal de n*n, $\Sigma$ una matriz de m*n "casi" diagonal salvo por el excedente de dimensiones y con diagonal $\sigma_1\ge\sigma_2\ge...\ge\sigma_{min(m,n)}\ge0 $

### 4. ¿Qué es la diagonalización de una matriz y que representan los eigenvectores?

La diagonalización de una matriz $A$ consiste en encontrar una matriz diagonal $D$ tal que $D=B^{-1}AB$,
donde $B^{-1}$ es la inversa de una matriz $B$.

Un eigenvector $v$ es una vector que cumple con la caracteristica de que dada una matriz A de n*n $Av=\lambda v$, con $v\ne0$

### 5. ¿Intuitivamente qué son los eigenventores?

El eigenvector de una transformación lineal es un vector (diferente de cero) para el cual la dirección no cambia cuando se le aplica la transformación lineal, es decir $Av$ y $\lambda v$ son paralelos y escalados por el factor $\lambda$.

### 6. ¿Cómo interpretas las descomposición en valores singulares como una composición de tres tipos de transformaciones lineales simples?

La descomposición de valores singulares dice que toda transformación lineal es una rotación ($V^t$) + una redimensionalización de los ejes canónicos ($\Sigma V^t$) y + una rotación ($U\Sigma V^t$)

### 7. ¿Qué relación hay entre la descomposición en valores singulares y la diagonalización?

 Tenemos que $A=U\Sigma V^t$

Por lo que $A^tA=(U\Sigma V^t)^t(U\Sigma V^t)=V\Sigma^tU^tU\Sigma V^t=V\Sigma^t\Sigma V^t$; como $U$ es ortogonal: $U^tU=I$

Además, $D=\Sigma^t\Sigma$ es una matriz con los valores $\sigma_1^2$, $\sigma_2^2$, ... en la diagonal, y los demás valores son cero. A esto se le conoce como diagonalización.

Entonces $A^tA=VDV^t=VDV^{-1}$; en el caso de matrices ortogonales, la inversa coincide con la transpuesta.

### 2. Aplicación a la pseudoinversa y sistemas de equaciones

```python
class SVD:
    "Constructor"    
    def __init__(self):
        pass
    
    def pseudoinversa(mat): 
        "Devuelve la pseudoinversa a partir de la SVD"
        rowsA, colsA = mat.shape
        U, s, V_t = np.linalg.svd(mat)
        sigma= np.array([[0.0 for l in range(colsA)] for i in range(rowsA)])
        for i in range(min(rowsA,colsA)):
            sigma[i,i]=s[i]
        sigma_sqr=np.array([[0.0 for l in range(min(rowsA,colsA))] for i in range(min(rowsA,colsA))])
        for i in range(min(rowsA,colsA)):
            sigma_sqr[i,i]=s[i]
        sigma_sqr_inv=np.linalg.inv(sigma_sqr)
        sigma_inv= np.array([[0.0 for l in range(rowsA)] for i in range(colsA)])
        for i in range(min(rowsA,colsA)):
            sigma_inv[i,i]=sigma_sqr_inv[i,i]        
        U_t=U.transpose()
        V=V_t.transpose()
        Vsigma_inv=np.matmul(V,sigma_inv)        
        pseudoA=np.matmul(Vsigma_inv,U_t)
        return pseudoA
    
    def sistema(mat,b):
        "Resuelve el sistema Ax=b con pseudoinversa"
        "Devuelve la pseudoinversa a partir de la SVD"
        rowsA, colsA = mat.shape
        U, s, V_t = np.linalg.svd(mat)
        sigma= np.array([[0.0 for l in range(colsA)] for i in range(rowsA)])
        for i in range(min(rowsA,colsA)):
            sigma[i,i]=s[i]
        sigma_sqr=np.array([[0.0 for l in range(min(rowsA,colsA))] for i in range(min(rowsA,colsA))])
        for i in range(min(rowsA,colsA)):
            sigma_sqr[i,i]=s[i]
        sigma_sqr_inv=np.linalg.inv(sigma_sqr)
        sigma_inv= np.array([[0.0 for l in range(rowsA)] for i in range(colsA)])
        for i in range(min(rowsA,colsA)):
            sigma_inv[i,i]=sigma_sqr_inv[i,i]        
        U_t=U.transpose()
        V=V_t.transpose()
        Vsigma_inv=np.matmul(V,sigma_inv)        
        pseudoA=np.matmul(Vsigma_inv,U_t)        
        x=np.matmul(pseudoA,b)       
        return x
```

```python
SVD.pseudoinversa(np.array([[2,4,6], [4,5,6],[3,1,-2]]))
```

```python
SVD.sistema(np.array([[2,4,6], [4,5,6],[3,1,-2]]),np.array([18,24,4]))
```

La $Im(A)=\{(x_1, x_2) | x_1+x_2=b_1\}= \Re^2$

$b$ no puede estar en la $Im(A)$, porque $b=\begin{bmatrix}b_1\\b_2\end{bmatrix}=\begin{bmatrix}x_1+x_2\\0\end{bmatrix}$, entonces no puede darse el caso en que $Im(A)$, porque $b=\begin{bmatrix}1\\1\end{bmatrix}$

La solución es infinita en $\Re^2$, ya que hay solución para toda $x_1 y x_2$ tales que $b_1=x_1+x_2$
